```python
import os
text_files_path = '/content/drive/My Drive/txt'  # Đường dẫn đến thư mục chứa tệp .txt
text_files = [os.path.join(text_files_path, f) for f in os.listdir(text_files_path) if f.endswith('.txt')]
```

```python
# Định nghĩa thư mục đầu ra cho âm thanh
output_wav_folder = '/content/drive/My Drive/output_wav'

# Tạo thư mục nếu chưa tồn tại
if not os.path.exists(output_wav_folder):
    os.makedirs(output_wav_folder)

# Dùng một vòng lặp để đọc từng tệp văn bản và chuyển đổi thành âm thanh
for text_file in text_files:
    with open(text_file, 'r') as file:
        text = file.read()

    # Định nghĩa tên tệp đầu ra
    output_wav_file = os.path.join(output_wav_folder, os.path.basename(text_file).replace('.txt', '.wav'))

    # Chuyển đổi văn bản thành âm thanh
    !python3 -m vietTTS.synthesizer --lexicon-file assets/infore/lexicon.txt --text="{text}" --output="{output_wav_file}" --silence-duration 0.2
```
